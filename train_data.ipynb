# Machine learning

```python
from sklearn.model_selection import train_test_split 
from sklearn.ensemble import RandomForestRegressor
import xgboost as xgb  
from xgboost import  XGBRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import GridSearchCV
from sklearn import metrics

import warnings
warnings.filterwarnings("ignore")
```

```python
fltr.head()
```

```python
X = fltr.drop('rul',axis=1)
Y = fltr.rul
# xt = tsdf.drop("rul",axis=1)
# yt = tsdf.rul
X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.20,random_state=42)
```

```python
def ml_func (algoritm):
    #train and fit regression model
    model=algoritm()
    model.fit(X_train, y_train)

    # predict
    train_preds = model.predict(X_train)
    test_preds = model.predict(X_test)

    # evaluate
    train_error = np.sqrt(metrics.mean_squared_error(y_train ,train_preds))
    test_error = np.sqrt(metrics.mean_squared_error(y_test, test_preds))
    #report = classification_report(y_test, test_preds)

    print(str(algoritm))
    print("------------------------")
    print('RMSE on Train set:', np.sqrt(metrics.mean_squared_error(y_train ,train_preds)))
    print('RMSE on Test set:', np.sqrt(metrics.mean_squared_error(y_test, test_preds)))

    # store accuracy in a new dataframe
    score_logreg = [algoritm, train_error, test_error]
    models = pd.DataFrame([score_logreg])
```

```python
algoritm = [XGBRegressor,KNeighborsRegressor,RandomForestRegressor]
#running each model and print accuracy scores
for algoritm in algoritm:
    ml_func (algoritm)
```

```python
def grid_src(regressor, param_grid):
    param_grid=param_grid
    # instantiate the tuned random forest
    grid_search = GridSearchCV(regressor, param_grid, cv=3, n_jobs=4,verbose=3)

    # train the tuned random forest
    grid_search.fit(X_train, y_train)

    # print best estimator parameters found during the grid search
    print((str(regressor) + 'Best Parameters'))
    print("------------------------")
    print(grid_search.best_params_)
    return grid_search.best_params_
```

```python
param_grid_rf = {'n_estimators': [10,50, 100],
                  'max_depth': [5,7,10], 
                  'min_samples_split': [2, 3, 4],
                 'min_samples_leaf':[2,3,5],
                 'random_state':[0,42]
                 
                 }

rf_params=grid_src(RandomForestRegressor(),param_grid_rf)
```

```python
param_grid_xg = {'n_estimators': [10,50,100],
                 'booster': ['gbtree','gblinear'],
              'learning_rate': [0.01,0.05], 
              'max_depth': [3, 5, 10],
              'gamma': [0.0, 0.1, 0.2],
                }
xg_params = grid_src(XGBRegressor(), param_grid_xg)
```

# Run models with best parameters

```python
def run_model(model, X_train, y_train,X_test, y_test ):
    model.fit(X_train, y_train)

    # predict
    train_preds = model.predict(X_train)
    test_preds = model.predict(X_test)


    #print reports of the model accuracy
    print('Model Scores')
    print("------------------------")
    print("------------------------")
    print('RMSE on Train set:', np.sqrt(metrics.mean_squared_error(y_train ,train_preds)))
    print('RMSE on Test set:', np.sqrt(metrics.mean_squared_error(y_test, test_preds)))
```

```python
rf_model=RandomForestRegressor(n_estimators=100, 
                                  max_depth= 10, 
                                  min_samples_split= 2,
                                  min_samples_leaf = 5,
                                  random_state = 0
                                  )
                               
                               
run_model(rf_model, X_train, y_train,X_test, y_test)
```

```python
xg_model = XGBRegressor(booster= 'gbtree',
                         gamma=0.2,
                         learning_rate= 0.03,
                         max_depth= 7,
                        n_estimators= 100)
```
